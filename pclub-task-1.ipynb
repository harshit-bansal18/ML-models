**Linear Model with cubic loss function - |x-x'|^3**

```python
def linear_regression_model1(x_train, y_train, x_test, y_test, learning_rate=0.03, num_iterations=500):
    #initialize parameters
    m, n = x_train.shape
    m_test,_ = x_test.shape
    theta = np.zeros((n,1))
    for i in range(num_iterations):
        y = np.dot(x_train, theta)
        cost = (1/(3*m))*np.sum(np.power(np.abs(y-y_train), 3))
        cost_eff  = cost = (1/(3*m))*np.sum(np.power(y-y_train, 3))
        #update theta by gradient descent
        d1 = np.power(y-y_train,2)*x_train
        derv = (1/m)*np.reshape(np.sum(d1, axis=0), (n,1))
        if cost_eff > 0:
            theta = theta - learning_rate*(derv)
        else:
            theta = theta + learning_rate*(derv)
        if i%100 == 0:
            print("Cost after",i, "iteration:",cost)
            
    
    #predict on test dataset
    print(theta.shape)
    predictions = np.dot( x_test, theta)
    cost = (1/(3*m_test))*np.sum(np.power(np.abs(predictions-y_test), 3)) 
    #print("Predicted price on test dataset: ",predictions)
    print("Error in predictions: ",cost)
    l = plt.plot(y_test, 'r', predictions,'b')
    plt.xlabel('Test cases')
    plt.ylabel('Prices')
    plt.title('Comparsion btw Actual and Predicted prices')
    plt.legend(l, ['Actual Price', 'Predicted Price'], fancybox=True)
    plt.show()
    
```

**Polynomial Model with 4th degree loss function - |x-x'|^4**

```python
def poly_regression_model1(x_train, y_train, x_test, y_test, learning_rate=0.1, num_iterations=2000):
    #initialize parameters
    m, n = x_train.shape
    m_test,_ = x_test.shape
    theta = np.zeros((n,1))
    for i in range(num_iterations):
        y = np.dot(x_train, theta)
        cost = (1/(4*m))*np.sum(np.power(np.abs(y-y_train), 4))
        #update theta by gradient descent
        d1 = np.power(y-y_train,3)*x_train
        derv = (1/m)*np.reshape(np.sum(d1, axis=0), (n,1))
        theta = theta - learning_rate*(derv)
        if i%100 == 0:
            print("Cost after",i, "iteration:",cost)
    
    #predict on test dataset
    print(theta.shape)
    predictions = np.dot( x_test, theta)
    cost = (1/4*(m_test))*np.sum(np.power(np.abs(predictions-y_test), 4)) 
    #print("Predicted price on test dataset: ",predictions)
    print("Error in predictions: ",cost)
    l = plt.plot(y_test, 'r', predictions,'b')
    plt.xlabel('Test cases')
    plt.ylabel('Prices')
    plt.title('Comparsion btw Actual and Predicted prices')
    plt.legend(l, ['Actual Price', 'Predicted Price'], fancybox=True)
    plt.show()
```

```python
data = pd.read_csv('/kaggle/input/usahousing/USA_Housing.csv')

```

```python
def process_linear_data(data):
    m = len(data)
    income = np.reshape(np.array(data['Avg. Area Income']), (1,m))
    income = income/np.std(income)
    age = np.reshape(np.array(data['Avg. Area House Age']), (1,m))
    age = age/np.std(age)
    rooms = np.reshape(np.array(data['Avg. Area Number of Rooms']), (1,m))

    rooms = rooms/np.std(rooms)
    beds = np.reshape(np.array(data['Avg. Area Number of Bedrooms']), (1,m))

    beds = rooms/np.std(beds)
    pop = np.reshape(np.array(data['Area Population']), (1,m))

    pop = pop/np.std(pop)
    feat = np.ones((1,len(data)))
    X = np.concatenate((feat, income/5, age/5, rooms/5, beds/5, pop/5), axis = 0).T
    x_train = X[0:4000,:]
    x_test = X[4000:5000,:]
    Y = np.reshape(np.array(data['Price']/1000000), (1,m)).T
    y_train = Y[0:4000,:]
    y_test = Y[4000:5000,:]
    return (x_train, x_test, y_train, y_test)
```

```python
def process_poly_data(data):
    m = len(data)
    income = np.reshape(np.array(data['Avg. Area Income']), (1,m))
    income = income/np.std(income)
    age = np.reshape(np.array(data['Avg. Area House Age']), (1,m))
    age = age/np.std(age)
    rooms = np.reshape(np.array(data['Avg. Area Number of Rooms']), (1,m))
    rooms = np.power(rooms, 2)
    rooms = rooms/np.std(rooms)
    beds = np.reshape(np.array(data['Avg. Area Number of Bedrooms']), (1,m))
    beds = np.power(beds, 2)
    beds = rooms/np.std(beds)
    pop = np.reshape(np.array(data['Area Population']), (1,m))
    
    pop = pop/np.std(pop)
    feat = np.ones((1,len(data)))
    X = np.concatenate((feat, income/5, age/5, rooms/5, beds, pop/5), axis = 0).T
    x_train = X[0:4000,:]
    x_test = X[4000:5000,:]
    Y = np.reshape(np.array(data['Price']/1000000), (1,m)).T
    y_train = Y[0:4000,:]
    y_test = Y[4000:5000,:]
    return (x_train, x_test, y_train, y_test)
```

```python
x_train, x_test, y_train, y_test = process_linear_data(data)
print(x_train, y_train)
print(y_train.shape)
```

```python
px_train, px_test, py_train, py_test = process_poly_data(data)
print(px_train, py_train)
```

```python
linear_regression_model1(x_train, y_train, x_test, y_test, learning_rate= 0.18, num_iterations=800)
```
